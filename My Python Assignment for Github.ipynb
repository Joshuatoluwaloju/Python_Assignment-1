Write a Python program to get the largest number from a list

```python
# Define a list of numbers
numbers = [15, 5, 3, 48, 22, 99, 97]

# Use the max() function to find the largest number
largest = max(numbers)

# Print the result
print("The largest number in the list is:", largest)
```

Write a Python program to get the smallest number from a list

```python
# Define a list of numbers
numbers = [20, 5, 77, 44, 22, 40, 1]

# Get the smallest number using min()
smallest = min(numbers)

# Print the result
print("The smallest number in the list is:", smallest)
```

Write a Python script to add a key to a dictionary.

```python
# Initial dictionary
person = {
    "name": "Joshua",
    "age": 20
}

# Add a new key-value pair
person["city"] = "Kwara"

# Print updated dictionary
print("Updated dictionary:", person)
```

Write a Python script to merge two Python dictionaries

```python
# Define two dictionaries
dict1 = {"name": "Joshua", "age": 20}
dict2 = {"city": "Kwara", "country": "Nigeria"}

# Merge dict2 into dict1
dict1.update(dict2)

# Print the merged dictionary
print("Merged dictionary:", dict1)
```

Write a Python program to remove a key from a dictionary

```python
# Define a dictionary
person = {
    "name": "Joshua",
    "age": 20,
    "city": "Kwara"
}

# Remove the 'city' key
removed_value = person.pop("city")

# Print the updated dictionary and removed value
print("Updated dictionary:", person)
print("Removed value:", removed_value)
```

Write a Python program to convert a list to a tuple.

```python
# Define a list
my_list = [2, 3, 4, 5, 6]

# Convert the list to a tuple
my_tuple = tuple(my_list)

# Print the result
print("Original list:", my_list)
print("Converted tuple:", my_tuple)
```
